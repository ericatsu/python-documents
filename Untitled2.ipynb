```python
import numpy as np;
import matplotlib.pyplot as plt;
f=lambda x:2*x**3-3*x**2+5;
fprime=lambda x:6*x**2-6*x;
#Create input array and it to decimal places
x=np.around(np.arange(-1.0,2.0,0.01),2)
#x=np.round_(x, decimals = 2)

y=f(x)
yprime=fprime(x)
yprimesign=np.sign(yprime);
sgnchg=yprimesign[:-1]*yprimesign[1:]

t=np.where(sgnchg<=0)

#x_=x[:-1]
#y_=y[:-1]

turningX=x[t].reshape(-1,1)
turningY=y[t].reshape(-1,1)
turningpoints = np.concatenate((turningX, turningY), axis=1)
plt.plot(x,y,color='red');
plt.plot(turningpoints[:,0],turningpoints[:,1], marker="X", markersize=15);
```
